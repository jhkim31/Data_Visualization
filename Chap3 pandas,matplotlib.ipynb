# matplotlib의캔버스,axes객체

그래프를 그리기 위해서는 캔버스와 Axes객체가 필요하다.  
해당 객체를 만들기 위해 다음과 같은 방법들이 있음
### 1. plt.plot(), plt.scatter()등의 메소드들은 자동으로 캔버스와 Axes를 만들어 사용한다(편한방법).  
이 방법으로 생성하게 되면 다루게 되는 캔버스, Axes가 하나인것 같다.(정확하지는 않음)  

### 2. 직접 캔버스와, Axes객체를 만든다
```python
fig = plt.figure()
ax = fig.add_subplot()
```
위와같이 사용하여 캔버스를 만든 다음 해당 캔버스에 axes객체를 추가하는 방법을 사용함.

## plt에 대해 설정과, Axes에 대한 설정
plt.plot()의 경우 하나의 Axes만 다루기 때문에 타이틀 설정, x,y축 설정등이 하나의 객체에 적용된다. 하지만 Axes를 따로 지정한 경우 여러개의 Axes가 있기 떄문에 각 Axes에 대해 적용을 해야함.  
**이때 plt와, Axes에 대해 적용하는 방법이 다르다.**  

ex) xlabel을 추가할때 
* 1. 의 방법의 경우 plt.xlabel() 의 방법으로 추가하지만,  
* 2. 의 방법의 경우 ax.set_xlabel()의 방법으로 추가하게 된다.  

대부분의 경우 Axes객체에 직접 하게되는 경우 `plt.***`메소드에 `set_` 을 붙여 사용하게 된다
* plt.xlabel() => ax.set_xlabel()
* plt.title() => ax.set_title

**하지만 모든 설정이 다른건 아님!!**
* plt.grid(), ax.grid() : 그리드 표현 여부를 설정한다. (동일함)




````python
plt.plot([1,2,3,4,5])
plt.title('asdf')
plt.grid(True)
plt.show()
````

````python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot([1,2,3,4,5])
ax.set_title('asdf')
ax.grid(True)
plt.show()
````

# plt공통메소드

* `axhline(y=0, xmin=0, xmax=1, **kwargs)` : y는 수평선이 그려 지는 y축의 좌표를 의미합니다.
    * 예시) plt.axhline(y=idx, color='c', linewidth=1, linestyle='dashed') 반대 함수는 plt.axvline()입니다.
* `axvline(x=0, ymin=0, ymax=1, **kwargs)` 
    * plt.axvline(x=1, color='blue', linewidth=2, linestyle='dotted') 
    * plt.axvline(x=2, color='red', linewidth=2, linestyle='solid')
* grid(boolean) : 그리드를 표시할 것인지의 Boolean 값을 지정합니다.
* rcdefaults() : Matplotlib 스타일에 대한 파라미터를 초기 값으로 되돌립니다.
* show() : 그래프 객체를 화면에 보여 줍니다.
* title(sometitle) : sometitle이라는 문장을 그래프의 제목으로 설정합니다.
* xlabel(str) : x 축에 str이라는 라벨(문구)을 설정합니다.
* xlim([하한값, 상한값]) : x 축의 상한/하한 값을 설정합니다.
* xticks(리스트) : x 축의 눈금 라벨에 대한 개수 설정을 합니다. 
    * ex) plt.xticks(np.arange(1, 11))
* ylabel(str) : y 축에 str이라는 라벨을 설정합니다.
* ylim([하한값, 상한값] : y 축의 상한/하한 값을 설정합니다.
* yticks(리스트) : y 축의 눈금 라벨에 대한 개수 설정을 합니다. 
    * ex) plt.yticks(np.arange(ymin, ymax + 1, 1))

# 이미지관련메소드
사용 형식
```
plt.savefig(filename, dpi=400, bbox_inches='tight')
```
* fname : 파일 경로나 파이썬의 파일과 유사한 객체를 나타내는 문자열. 저장되는 포맷은 파일의 확장자를 통하여 결정이 됩니다.
* dpi : Figure의 해상도 dpi(기본 값) 100)를 지정합니다. dpi는 dots per inch를 의미합니다.
* facecolor, edge color : 서브 플롯 바깥 배경 색상, 기본 값) w(흰색)입니다.
* format : 명시적인 파일 포맷('png', 'pdf', 'svg' 등등)을 지정합니다.
* bbox_inches : Figure에서 저장할 부분을 지정하는 옵션입니다.
    * 'tight'는 Figure의 둘레의 비어 있는 공간을 모두 제거합니다.

# pyplot(plt)

# 단일꺾은선그래프

* plt.plot(data, color, marker)

# 이중축꺾은선그래프

* [plt.subplots(nrows=1, ncols=1, *)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib-pyplot-subplots) : figure와 Axes를 리턴한다 
    * 만약 nrows, ncols가 2 이상이면 fig와 Axes배열을 리턴한다.
    ```
    nrows = 3, ncols = 3
    
    [[<AxesSubplot:> <AxesSubplot:> <AxesSubplot:>]
     [<AxesSubplot:> <AxesSubplot:> <AxesSubplot:>]
     [<AxesSubplot:> <AxesSubplot:> <AxesSubplot:>]]
    ```
* ax.twinx() : 같은 x축을 공유하는 플롯
* ax.set_ylabel(str, color) : 해당 Axes의 y축라벨을 설정함
* ax.tick_params(axis, labelcolor) : 지정한 축의 설정 변경

````python
tips_frame = pd.read_csv('./data/seaborn-data/tips.csv')

print("\033[31mtips_frame.head()\033[0m")
print(tips_frame.head())
print()

print("\033[31mtips_frame : \033[0m")
print(tips_frame.info())
print()

print("\033[31mtips_frame.describe()\033[0m")
print(tips_frame.describe())
print()

fig, axes = plt.subplots(2,2)
ax1 = axes[0,0]


ax1.plot(tips_frame.loc[:, ['total_bill']], color='tab:red')
ax1.set_ylabel('total_bill', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.plot(tips_frame.loc[:, ['tip']], color='tab:blue')
ax2.set_ylabel('tip', color = 'tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

fig.tight_layout()
plt.show()
````

````python
fig = plt.figure()
mpg = pd.read_csv('./data/ggplot/mpg.csv')

print("\033[31mdrv가 f(전륜)인 행만 출력\033[0m")
print(mpg.drv == 'f')
print()
# 전륜인 차들의 displ, hwy 값
x1 = mpg.loc[mpg.drv == 'f', 'displ']
y1 = mpg.loc[mpg.drv == 'f', 'hwy']

x2 = mpg.loc[mpg.drv == '4', 'displ']
y2 = mpg.loc[mpg.drv == '4', 'hwy']

x3 = mpg.loc[mpg.drv == 'r', 'displ']
y3 = mpg.loc[mpg.drv == 'r', 'hwy']


plt.plot(x1, y1, color='r', marker='o', linestyle='None')
plt.plot(x2, y2, color='g', marker='o', linestyle='None')
plt.plot(x3, y3, color='b', marker='o', linestyle='None')
plt.legend(['전륜',  '사륜', '후륜'])
plt.xlabel('엔진 크기')
plt.ylabel('주행 마일 수')
plt.show()
````

## 산점도 그래프와 히스토그램

* plt.GridSpec(r,c, hspace, wspace) : r,c에 따른 그리드를 설정한다.
    * 해당 객체는 그리드 객체는 다음과 같이 사용한다
        * grid[:-1, :-1] : 행 0 ~ 마지막 전 행까지, 열 0 ~ 마지막 전 열까지
        * grid[:-1, -1] : 행 0 ~ 마지막 전 행까지, 열 마지막 열
        * grid[3:5,1:4] : 행 3 ~ 4 행까지, 열 1 ~ 3열
        
* fig.add_subpot(position) : 캔버스에 Axes 객체를 추가함 포지션 방식은 여러가지다
    * 211 / 212, : 2행, 1열짜리 크기에 첫번째 객체(1행 1열)로 사용 / 2행, 1열짜리 크기에 두번째 객체(2행 1열)로 사용
    * grid[0:1, :] / grid[1:2, :] : 위와 동일한 포지션이다.
    

* Axes.hist(data, orientation, color) : 데이터를 통한 히스토그램을 설정한다.
    * data 예시) [1,2,2,3,3,3,3] 이라면 1은 하나, 2는 둘, 3은 네개가 있는 히스토그램이 그려진다
    * orientation : 방향 설정 default = 'vertical' ('horizontal')

* Axes.invert_yaxis() : y축을 기준으로 invert(뒤집는다)한다.
    
    

````python
fig = plt.figure(figsize=(16,10))
grid = plt.GridSpec(4,4, hspace=0.5, wspace=0.5)

ax_main = fig.add_subplot(grid[:-1, :-1])
ax_right = fig.add_subplot(grid[:-1, -1])
ax_bottom = fig.add_subplot(grid[-1, :-1], xticklabels=[], yticklabels=[])



mpg = pd.read_csv('./data/ggplot/mpg.csv')

print("\033[31mdrv가 f(전륜)인 행만 출력\033[0m")
print(mpg.drv == 'f')
print()
# 전륜인 차들의 displ, hwy 값
x1 = mpg.loc[mpg.drv == 'f', 'displ']
y1 = mpg.loc[mpg.drv == 'f', 'hwy']

x2 = mpg.loc[mpg.drv == '4', 'displ']
y2 = mpg.loc[mpg.drv == '4', 'hwy']

x3 = mpg.loc[mpg.drv == 'r', 'displ']
y3 = mpg.loc[mpg.drv == 'r', 'hwy']


ax_main.plot(x1, y1, color='r', marker='o', linestyle='None')
ax_main.plot(x2, y2, color='g', marker='o', linestyle='None')
ax_main.plot(x3, y3, color='b', marker='o', linestyle='None')
ax_main.legend(['전륜',  '사륜', '후륜'])
ax_main.set_xlabel('엔진 크기')
ax_main.set_ylabel('주행 마일 수')
ax_main.grid(True)

ax_right.hist(mpg.hwy, orientation='horizontal', color='lightblue')

ax_bottom.hist(mpg.displ, orientation='vertical', color='lightpink')
ax_bottom.invert_yaxis()

plt.show()
````
