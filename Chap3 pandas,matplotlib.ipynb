# matplotlib의캔버스,axes객체

그래프를 그리기 위해서는 캔버스와 Axes객체가 필요하다.  
해당 객체를 만들기 위해 다음과 같은 방법들이 있음
### 1. plt.plot(), plt.scatter()등의 메소드들은 자동으로 캔버스와 Axes를 만들어 사용한다(편한방법).  
이 방법으로 생성하게 되면 다루게 되는 캔버스, Axes가 하나인것 같다.(정확하지는 않음)  

### 2. 직접 캔버스와, Axes객체를 만든다
```python
fig = plt.figure()
ax = fig.add_subplot()
```
위와같이 사용하여 캔버스를 만든 다음 해당 캔버스에 axes객체를 추가하는 방법을 사용함.

## plt에 대해 설정과, Axes에 대한 설정
plt.plot()의 경우 하나의 Axes만 다루기 때문에 타이틀 설정, x,y축 설정등이 하나의 객체에 적용된다. 하지만 Axes를 따로 지정한 경우 여러개의 Axes가 있기 떄문에 각 Axes에 대해 적용을 해야함.  
**이때 plt와, Axes에 대해 적용하는 방법이 다르다.**  

ex) xlabel을 추가할때 
* 1. 의 방법의 경우 plt.xlabel() 의 방법으로 추가하지만,  
* 2. 의 방법의 경우 ax.set_xlabel()의 방법으로 추가하게 된다.  

대부분의 경우 Axes객체에 직접 하게되는 경우 `plt.***`메소드에 `set_` 을 붙여 사용하게 된다
* plt.xlabel() => ax.set_xlabel()
* plt.title() => ax.set_title

**하지만 모든 설정이 다른건 아님!!**
* plt.grid(), ax.grid() : 그리드 표현 여부를 설정한다. (동일함)




````python
plt.plot([1,2,3,4,5])
plt.title('asdf')
plt.grid(True)
plt.show()
````

````python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot([1,2,3,4,5])
ax.set_title('asdf')
ax.grid(True)
plt.show()
````

# plt공통메소드 
## (괄호는 Axes 적용시 메소드가 다를때)

* `axhline(y=0, xmin=0, xmax=1, **kwargs)` : y는 수평선이 그려 지는 y축의 좌표를 의미합니다.
    * 예시) plt.axhline(y=idx, color='c', linewidth=1, linestyle='dashed') 반대 함수는 plt.axvline()입니다.
* `axvline(x=0, ymin=0, ymax=1, **kwargs)` 
    * plt.axvline(x=1, color='blue', linewidth=2, linestyle='dotted') 
    * plt.axvline(x=2, color='red', linewidth=2, linestyle='solid')
* grid(boolean) : 그리드를 표시할 것인지의 Boolean 값을 지정합니다.
* rcdefaults() : Matplotlib 스타일에 대한 파라미터를 초기 값으로 되돌립니다.
* show() : 그래프 객체를 화면에 보여 줍니다.
* title(sometitle) **(set_title)**: sometitle이라는 문장을 그래프의 제목으로 설정합니다.
* xlabel(str) **(set_xlabel)** : x 축에 str이라는 라벨(문구)을 설정합니다.
* xlim([하한값, 상한값]) **(set_xlim)** : x 축의 상한/하한 값을 설정합니다.
* xticks(리스트) **(set_xticks)** : x 축의 눈금 라벨에 대한 개수 설정을 합니다. 
    * ex) plt.xticks(np.arange(1, 11))
* ylabel(str) **(set_ylabel)** : y 축에 str이라는 라벨을 설정합니다.
* ylim([하한값, 상한값] **(set_ylim)** : y 축의 상한/하한 값을 설정합니다.
* yticks(리스트) **(set_yticks)** : y 축의 눈금 라벨에 대한 개수 설정을 합니다. 
    * ex) plt.yticks(np.arange(ymin, ymax + 1, 1))

# 이미지관련메소드
사용 형식
```
plt.savefig(filename, dpi=400, bbox_inches='tight')
```
* fname : 파일 경로나 파이썬의 파일과 유사한 객체를 나타내는 문자열. 저장되는 포맷은 파일의 확장자를 통하여 결정이 됩니다.
* dpi : Figure의 해상도 dpi(기본 값) 100)를 지정합니다. dpi는 dots per inch를 의미합니다.
* facecolor, edge color : 서브 플롯 바깥 배경 색상, 기본 값) w(흰색)입니다.
* format : 명시적인 파일 포맷('png', 'pdf', 'svg' 등등)을 지정합니다.
* bbox_inches : Figure에서 저장할 부분을 지정하는 옵션입니다.
    * 'tight'는 Figure의 둘레의 비어 있는 공간을 모두 제거합니다.

## 산점도 그래프와 히스토그램

* plt.GridSpec(r,c, hspace, wspace) : r,c에 따른 그리드를 설정한다.
    * 해당 객체는 그리드 객체는 다음과 같이 사용한다
        * grid[:-1, :-1] : 행 0 ~ 마지막 전 행까지, 열 0 ~ 마지막 전 열까지
        * grid[:-1, -1] : 행 0 ~ 마지막 전 행까지, 열 마지막 열
        * grid[3:5,1:4] : 행 3 ~ 4 행까지, 열 1 ~ 3열
        
* fig.add_subpot(position) : 캔버스에 Axes 객체를 추가함 포지션 방식은 여러가지다
    * 211 / 212, : 2행, 1열짜리 크기에 첫번째 객체(1행 1열)로 사용 / 2행, 1열짜리 크기에 두번째 객체(2행 1열)로 사용
    * grid[0:1, :] / grid[1:2, :] : 위와 동일한 포지션이다.
    

* Axes.hist(data, orientation, color) : 데이터를 통한 히스토그램을 설정한다.
    * data 예시) [1,2,2,3,3,3,3] 이라면 1은 하나, 2는 둘, 3은 네개가 있는 히스토그램이 그려진다
    * orientation : 방향 설정 default = 'vertical' ('horizontal')

* Axes.invert_yaxis() : y축을 기준으로 invert(뒤집는다)한다.
    
    

````python
fig = plt.figure(figsize=(16,10))
grid = plt.GridSpec(4,4, hspace=0.5, wspace=0.5)

ax_main = fig.add_subplot(grid[:-1, :-1])
ax_right = fig.add_subplot(grid[:-1, -1])
ax_bottom = fig.add_subplot(grid[-1, :-1], xticklabels=[], yticklabels=[])



mpg = pd.read_csv('./data/ggplot/mpg.csv')

print("\033[31mdrv가 f(전륜)인 행만 출력\033[0m")
print(mpg.drv == 'f')
print()
# 전륜인 차들의 displ, hwy 값
x1 = mpg.loc[mpg.drv == 'f', 'displ']
y1 = mpg.loc[mpg.drv == 'f', 'hwy']

x2 = mpg.loc[mpg.drv == '4', 'displ']
y2 = mpg.loc[mpg.drv == '4', 'hwy']

x3 = mpg.loc[mpg.drv == 'r', 'displ']
y3 = mpg.loc[mpg.drv == 'r', 'hwy']


ax_main.plot(x1, y1, color='r', marker='o', linestyle='None')
ax_main.plot(x2, y2, color='g', marker='o', linestyle='None')
ax_main.plot(x3, y3, color='b', marker='o', linestyle='None')
ax_main.legend(['전륜',  '사륜', '후륜'])
ax_main.set_xlabel('엔진 크기')
ax_main.set_ylabel('주행 마일 수')
ax_main.grid(True)

ax_right.hist(mpg.hwy, orientation='horizontal', color='lightblue')

ax_bottom.hist(mpg.displ, orientation='vertical', color='lightpink')
ax_bottom.invert_yaxis()

plt.show()
````

# 버블 차트

````python
diamonds = pd.read_csv('./data/ggplot/diamonds.csv')
diamonds = diamonds.sample(frac = 5e-3)

print("\033[31mDataFrame : 0.5% 만큼 샘플링된 diamonds 데이터\033[0m")
print(diamonds)
print()


xdata = diamonds.price
ydata = diamonds.depth
table = diamonds.table  #원의 크기

cut_list = diamonds.cut.unique()
clist = ['r', 'g', 'b', 'y', 'm']
print("\033[31mcut list\033[0m")
print(cut_list)
print()

cut_dict = {cut_list[i]:clist[i] for i in range(len(cut_list))}
print("\033[31mcut_dict \033[0m")
print(cut_dict)
print()

def reduce_cut(cut):
    return cut_dict[cut]
diamonds['newcut'] = diamonds.cut.apply(reduce_cut)
newcut = diamonds.newcut
newtable = diamonds.table.apply(lambda x : (x - 55) * 20)

print("\033[31mnewcut\033[0m")
print(newcut)
print()

print("\033[31mnewtable\033[0m")
print(newtable)
print()

fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(111)
ax1.scatter(x = xdata, y = ydata, s = newtable, c = newcut)
ax1.legend()
plt.show()
````

# 바 차트

## 기본 바차트
* ax1, plt.bar(x,height,color, ...) : 바차트를 만든다
    * x : x축 데이터
    * height : value
    * color : 색 데이터 (만약 x축 데이터보다 색이 부족하면 색을 반복하여 채운다)

* ax1, plt.text(x,y,s, horizontalalignment) : 텍스트를 넣는다
    * x : x position
    * y : y position
    * s : 넣으려는 string
    * horizontalalignment : 위치에 대해 텍스트 시작점을 잡는것 같음 
        * (center로 설정시 텍스트의 중앙이 해당 포지션, default는 왼쪽 끝이 해당 포지션)

````python
filesrc = './data/주요발생국가주간동향(4월2째주).csv' 
data = pd.read_csv(filesrc, index_col='국가')

xdata = data.index
ydata = data['4월06일']

print("\033[31mydata\033[0m")
print(ydata)
print()

fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot()
ax1.bar(xdata, ydata, color=['r', 'g', 'b'])
sum = ydata.sum()
ratio = list(map(lambda y : (y / sum) * 100, ydata))

for i in range(len(ydata)):
    y = ydata[i]
    rat = ratio[i]
    plt.text(x=i, y=y + 10000, s=f'{y}건', horizontalalignment='center')
    plt.text(x=i, y=y // 2, s=f'{rat:.2f}%', horizontalalignment='center')
    

ax1.axhline(y = ydata.mean(), color='red', linewidth=2, linestyle='dashed')
ax1.text(x = len(xdata) - 1, y = ydata.mean() + 2000, s = f'평균{ydata.mean()}건')
plt.show()
````

## 하나의 축에 여러개 바차트

* DataFrame.plot(kind, ax) : 현재 DataFrame 객체를 사용하여 플롯을 만든다.
    * 기본적으로 DataFrame의 **index를 x축으로 지정**한다.
    * kind : 종류 (default = plot)

## 가로 누적 바 차트

원리는 다음과 같음.  
.barh 메소드에는 left라는 파라미터가 있다. 이 파라미터는 데이터의 시작 지점을 설정할 수 있는데  
ex) left = 0, width = 10 이라고 차트를 그리고 이후 left=10, width=20이라고 차트를 그리면 0 ~ 10, 20 ~ 30 까지 차트가 그려진다.  
이를 잘 쓰기 위해 numpy의 cumsum()을 사용한다.  

* np.cumsum() : 배열의 누적합을 계산함 [1,2,3,4] => [1,3,6,10]

````python
filesrc = './data/주요발생국가주간동향(4월2째주).csv' 
data = pd.read_csv(filesrc, index_col='국가')
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot()

country = ['미국', '스페인', '독일', '이란']

data = data.loc[country]

labels = list(data.index)
data = data.values
data_cum = data.cumsum(axis=1)


print("\033[31mdata_cum\033[0m")
print(data_cum)
print()


for i in range(7):
    if i == 0:
        ax1.barh(labels, values[:, i])
    else:
        ax1.barh(labels, values[:, i], left = data_cum[:, i - 1])



````

## autopct 함수설정

pie의 파라미터 autopct에 함수를 설정할 수 있음 기본 파라미터로 퍼센테이지(float)를 던져준다.


````python
filesrc = './data/주요발생국가주간동향(4월2째주).csv' 
data = pd.read_csv(filesrc, index_col='국가')

country = ['독일', '프랑스', '중국', '영국']
data = data.loc[country]
chartdata = data['4월06일']

print("\033[31mchartdata\033[0m")
print(chartdata)
print()

def getLabelFormat(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.2f}%\n({:d} 명)".format(pct, absolute)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot()

ax.pie(chartdata, labels=country, shadow=True, explode=(0, 0.05, 0, 0),  autopct=lambda pct: getLabelFormat(pct, chartdata), startangle=90, counterclock=False)

plt.grid(True)
plt.legend(loc=3)
plt.show()
````

## 계층 파이 그래프

뭔 특별한 방법이 있는줄 알았는데,, 그냥 크기가 다른 두개의 파이 차트를 그리는것 뿐임 ㅋ.  
계층 파이 그래프를 그릴때 사용되는 파라미터  
* wedggeprops : 도넛형으로 만들어야 되므로 width를 적절하게 줘야하고 edgecolor를 잘 줘야 구분이 잘간다.
* radius : 도넛의 반지름을 설정하는 것으로 안에 들어가는 작은 원의경우 radius를 좀 더 작게 줘야한다. (outer - outer.wedgeprops.width)
* pctdistance : 퍼센테이지를 출력하는 위치를 설정한다 (0 ~ 1 이며 1에 가까울수록 중앙에서 더 멀리 쓰여짐)

````python
filesrc = './data/주요발생국가주간동향(4월2째주).csv' 
data = pd.read_csv(filesrc, index_col='국가')

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot()

country = ['독일', '프랑스', '중국', '영국', '이탈리아']
inner_data = data.loc[country, ['4월06일', '4월07일']].values

outer_data = inner_data.sum(axis=1) # 누계

ax.pie(outer_data, radius=1, wedgeprops=dict(width=0.3, edgecolor='w'), labels=country, autopct='%.2f%%', pctdistance=0.85)
ax.pie(inner_data.flatten(), radius=0.7, wedgeprops=dict(width=0.3, edgecolor='w'), autopct='%.2f%%', pctdistance=0.75)
plt.show()
````

## Rectangular, notched

* notched는 중앙값 부분이 움푹 파여있어 더 찾기 쉬움

* axes.boxplot(data, labels, patch_artist, vert)
    * data : 데이터 
    * labels : x 라벨로써 데이터의 행만큼 있어야함
    * patch_artist : 색깔을 칠하는지에 대한 여부
    * vert : 세로 정렬이라는데 뭔지 모르겠음 ㅋ
    * notch : 노치로 할지에 대한 여부

* boxplot()의 반환값 .set_facecolor() : 박스 플롯의 색을 변경한다.

* axes.violinplot(data, showmeans, showmedians)
    * data : 데이터
    * showmeans : 평균값 보여줌 여부
    * showmedians : 중앙값 보여줌 여부
    * 바이올린 플롯은 labels를 지정할 수 없어서 다음과 같이 설정한다.
        * 1. axes자체에 set_xticks를 통해 xticks를 설정해주고
        * 2. 설정된 xtick들을 라벨로 바꿔(치환) 시켜준다.

````python
filename = './data/tips.csv'
myframe = pd.read_csv(filename, encoding='utf-8', index_col=0)

print(myframe[myframe.time == 'Dinner'])

frame1 = myframe.loc[myframe['time'] == 'Dinner', 'total_bill']
frame2 = myframe.loc[myframe['time'] == 'Lunch', 'total_bill']

labels = ['Dinner', 'Lunch']

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6), dpi=500)
ax1 = axes[0]
ax2 = axes[1]
ax3 = axes[2]

box1 = ax1.boxplot([frame1, frame2], labels=labels, patch_artist=True) 
box1['boxes'][0].set_facecolor('lightblue')
box1['boxes'][1].set_facecolor('lightpink')
ax1.yaxis.grid(True)

ax2.boxplot(frame1, labels=['Dinner'], notch=True)  
ax2.yaxis.grid(True)

ax3.violinplot([frame1, frame2], showmeans=True, showmedians=True)
ax3.set_xticks([1,2])
ax3.set(xticklabels=labels)

plt.show()
````
